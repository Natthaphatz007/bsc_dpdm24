```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score

# Step 1: Load the data
train_data = train_set  # Replace with your actual training data
test_data = x_test      # Replace with your actual test data

# Step 2: Separate features and target
X_train = train_data.drop(columns=['Thal_Group'])
y_train = train_data['Thal_Group']

# Step 3: Split train_data into training and validation sets
X_train_split, X_val_split, y_train_split, y_val_split = train_test_split(
    X_train, y_train, test_size=0.2, random_state=42
)

# Step 4: Identify numeric and non-numeric columns
numeric_cols = X_train.select_dtypes(include=['float64', 'int64']).columns
categorical_cols = X_train.select_dtypes(include=['object']).columns

# Step 5: Preprocessing pipeline
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean'))
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_cols),
        ('cat', categorical_transformer, categorical_cols)
    ])

# Step 6: Encode the target variable
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train_split)  # Fit on y_train_split
y_val_encoded = label_encoder.transform(y_val_split)          # Transform y_val_split

# Step 7: Create the model pipeline
model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(random_state=42))
])

# Step 8: Train the model on the training split
model.fit(X_train_split, y_train_encoded)

# Step 9: Make predictions on the validation split
y_val_pred_encoded = model.predict(X_val_split)

# Step 10: Calculate accuracy on the validation set
accuracy = accuracy_score(y_val_encoded, y_val_pred_encoded)
print(f'Validation Accuracy: {accuracy:.4f}')

# Step 11: Make predictions on X_test (if available)
if test_data is not None:
    y_pred_encoded = model.predict(test_data)
    y_pred = label_encoder.inverse_transform(y_pred_encoded)
    predictions_df = pd.DataFrame({'Thal_Group': y_pred})
    predictions_df.to_csv('y_pred_model.csv', index=False)
    print("Predictions saved to 'y_pred_model.csv'")
    print(predictions_df)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score

# Step 1: Load the data
train_data = train_set  # Replace with your actual training data
test_data = x_test      # Replace with your actual test data

# Separate features and target
X_train = train_data.drop(columns=['Thal_Group'])
y_train = train_data['Thal_Group']

# Use the actual X_test
X_test = test_data  # Ensure X_test has the same structure as X_train (same columns)

# Identify numeric and non-numeric columns
numeric_cols = X_train.select_dtypes(include=['float64', 'int64']).columns
categorical_cols = X_train.select_dtypes(include=['object']).columns

# Step 2: Preprocessing pipeline
# For numeric columns: Impute missing values with the mean
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean'))
])

# For categorical columns: Impute missing values with the most frequent value and one-hot encode
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Combine preprocessing steps
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_cols),
        ('cat', categorical_transformer, categorical_cols)
    ])

# Step 3: Encode the target variable (y_train)
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)

# Step 4: Create the model pipeline
model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(random_state=42))
])

# Step 5: Train the model
model.fit(X_train, y_train_encoded)

# Step 6: Make predictions on X_test
# Predict Thal_Group for X_test
y_pred_encoded = model.predict(X_test)

# Convert predictions back to original labels
y_pred = label_encoder.inverse_transform(y_pred_encoded)

# Step 7: Save predictions in the required format
predictions_df = pd.DataFrame({'Thal_Group': y_pred})
predictions_df.to_csv('y_pred_model.csv', index=False)

# Print the output
print(predictions_df)
```
