## Distance

````python
# prompt: build pd dataframe from [[1,2],[-1,4],[0,3],[-3,-2]]

import pandas as pd

data = [[1, 2], [-2, 4], [0, 3], [-3, -2]]
df = pd.DataFrame(data, columns=['X1', 'X2'])
df
````

````python
# prompt: draw scatter plot of the df,with X1 and X2 axis

import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame with columns 'X1' and 'X2'
plt.figure(figsize=(8, 6))  # Adjust figure size as needed
plt.scatter(df['X1'], df['X2'])
plt.plot([-3,1],[0,0],color='black', linewidth = 2)
plt.plot([0,0],[-2,4],color='black', linewidth = 2)
plt.xlabel('X1', fontweight='bold')
plt.ylabel('X2', fontweight='bold')
plt.title('Scatter Plot of X1 vs X2')
plt.grid(True)
plt.show()
````

````python
from scipy.spatial.distance import pdist, squareform

distance_matrix = squareform(pdist(df))

distance_matrix
````

````python
from scipy.spatial.distance import pdist, squareform

distance_matrix = squareform(pdist(df, 'cityblock')) # l1 norm is cityblock

distance_matrix
````

## Z-score

````python
# prompt: calculator z-score for df2

from scipy.stats import zscore

# Calculate z-scores for all columns in df2
df2_zscore = df2.apply(zscore)

df2_zscore
````

````python
# prompt: create the scatter plot for df2_zscore

min_value = df2_zscore.min()
max_value = df2_zscore.max()

print(min_value)
print(max_value)
plt.figure(figsize=(8, 6))
plt.scatter(df2_zscore['X1'], df2_zscore['X2'])
plt.xlabel('X1 (Z-score)', fontweight='bold')
plt.ylabel('X2 (Z-score)', fontweight='bold')
plt.title('Scatter Plot of Z-scores for X1 vs X2')
plt.grid(True)
plt.show()
````

````python
# prompt: do min-max normalization for the df2 for each column separately

# Min-Max normalization for each column of df2
df2_min_max = df2.copy()  # Create a copy to avoid modifying the original DataFrame

for column in df2.columns:
    min_val = df2[column].min()
    max_val = df2[column].max()
    df2_min_max[column] = (df2[column] - min_val) / (max_val - min_val)

df2_min_max
````
